```python
import numpy as np
import pandas as pd

rnames = ['user_id', 'movie_id', 'rating', 'timestamp'] 
ratings = pd.read_table('../data/ml-1m/ratings.dat', sep='::',engine='python', header=None, names=rnames)
```

```python
ratings
```

```python
Ys = np.mat(ratings)
Ys
```

```python
Nr = 6040 # Number of users
Nc = 3952 # Number of movies
Nk = 2

Ysize = Ys.shape[0]

A = np.mat(np.random.rand(Nr, Nk))
B = np.mat(np.random.rand(Nk, Nc))

EPOCH = 100
Eta = 0.1
eta = Eta
sample = 50000 # 50000 random rating will be chosen and error will be calculated only for those

#Obtain M matrix and Y matrix
M = np.mat(np.zeros((Nr, Nc)))
Y = np.mat(np.zeros((Nr, Nc)))

for i in range(Ysize):
    user = Ys[i,0]
    movie = Ys[i,1]
    M[user-1, movie-1] = 1;
    Y[user-1, movie-1] = Ys[i, 2]
 
Y
```

```python
for i in range(EPOCH):
    E = np.array(M)*np.array(Y - A*B)
    Err = np.sum(E*E)/np.sum(np.array(M))
    if i%10==0:
        print(Err)
    np.random.shuffle(Ys)
    
    for k in range(sample):
        #k = np.random.randint(0, Ysize)
        
        u = Ys[k,0] - 1
        m = Ys[k,1] - 1
        
        err = Ys[k,2] - A[u,:]*B[:,m]
        
        temp_A = A[u,:] + eta*err[0,0]*B[:,m].T
        B[:,m]   = B[:,m] + eta*err[0,0]*A[u,:].T
        A[u,:]   = temp_A
    
    eta = Eta*1./(i+1)

Yr = A*B
Yr
```

```python

result = np.zeros([20, 10])

Yp = A*B # This is a dummy matrix to be used while searching the max rated movies

# Find out the indeces of the maximum values in each row (corresponds to users)
# These indeces will be the max rated movies of this user
# In the assignment, we were asked to find out the 10 max rated movies of the first 20 users

for i in range(20):
    for j in range(10):
        idx = np.argmax(Yp[i, :])
        result[i, j] = idx
        Yp[i, idx] = 0 # reset to invalid rating to prevent this index to be found in further search


for i in range(20):
    u= str(i+1)
    print("User #" + u + "'s highest rating movies are:")
    for k in range(10):
        m = str(result[i, k])
        rate = round(Yr[i-1, result[i, k]], 3) # to get rid of unnecessary decimals
        rate = str(rate)
        print("\tMovie #" + m + " with rating " + rate)
```
